```python
# Se importa el archivo JSONL sobre información de venta de CDs y Vinylos a través de Amazon.

from google.colab import drive
drive.mount('/content/drive', force_remount=True)

import os
DIR = "/content/drive/My Drive/BigData/"
os.chdir(DIR)

import json
import pandas as pd

archivo_jsonl = 'meta_CDs_and_Vinyl.jsonl'
df = pd.read_json(archivo_jsonl, lines=True)
```

```python
# Selección de variables importantes para el análisis.

# Entendimiento general de las variables que componen la tabla.
df.info()
df.head(5)

# Con base en la información observada, se descartan laas variables sin información relevante o poco informadas.
```

```python
# Se seleccionan las variables: categoria musical, promedio de evaluación por disco y titulo de disco.
# Con estas tres variables, buscamos determinar los CDs mejor evaluados por las personas que compran en Amazon.

tb_rec = df[['categories','average_rating','title']]
tb_rec.info()
tb_rec.head(5)
```

```python
# Ordenamiento de variables para su correcto análisis.

tb_ordenada = tb_rec.sort_values(by=['average_rating'], ascending=[False])
print(tb_ordenada)
```

```python
# Se filtran solo los discos con la mejor evaluación, para generar una base de recomendaciones.

base_top = tb_ordenada[tb_ordenada['average_rating'] == 5.0]
base_top_ord = base_top.sort_values(by=['categories'], ascending=[False])
print(base_top_ord)
```

```python
# Seleccionamos los 3 primeros discos mejor evaluados por cada categoria.
# Para poder hacer esto, debemos generar una variable indice que cuente cada registro por categoria.

base_top_ord_ind = base_top_ord

base_top_ord_ind['categorias'] = base_top_ord_ind['categories']
base_top_ord_ind['categories'] = base_top_ord_ind['categories'].apply(lambda x: ' '.join(x))
base_top_ord_ind['Index'] = base_top_ord_ind.groupby('categories').cumcount() + 1

base_top_ord_ind.info()
```

```python
# Posteriormente seleccionamos solo los primeros 3 registros con la evaluación más alta por cada categoria.

top3 = base_top_ord_ind[base_top_ord_ind['Index'] <= 3]

Base_Recomendacion = top3[['categorias','title']]

Base_Recomendacion.head(15)
```
